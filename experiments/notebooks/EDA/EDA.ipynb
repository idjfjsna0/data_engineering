```python
numeric_cols = df.select_dtypes(include=['number']).columns.tolist() #  number-     
for name in numeric_cols:
    print(name)

outlier_summary = [] #   
for col in numeric_cols: #    
    ser = df[col]
    q1 = ser.quantile(0.25) # 25%    
    q3 = ser.quantile(0.75) # 75%    
    iqr = q3 - q1 #  , ,   
    lower = q1 - 1.5 * iqr #   
    upper = q3 + 1.5 * iqr #   
    outliers = ser[(ser < lower) | (ser > upper)]
    # ser < lower      
    #        ,    outliers
    pct = 100 * len(outliers) / len(ser) #   
    '''
              
    '''
    outlier_summary.append((col, len(outliers), round(pct,3), float(ser.min()), float(ser.max())))
    #       ,     

out_df = pd.DataFrame(outlier_summary, columns=['column','outlier_count','outlier_pct','min','max']).sort_values('outlier_pct', ascending=False)
#     ,     
display(out_df)
```

```python
#     (heatmap)
corr = df.select_dtypes(include=['number']).corr() #    
display(corr.round(3))

#   
plt.figure(figsize=(8,6))
plt.imshow(corr, aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.title('Correlation matrix (visual)')
plt.tight_layout()
plt.show()
```
